```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
```

```python
incendios_93_19_projected = gpd.read_file('data/incendios_93_19_clean_ETRS89.geojson')
```

```python
incendios_93_19_projected.geometry
```

```python
geo_03102_sample_df = gpd.read_file('data/municipios/datos_generados/03102/geo_03102_df.geojson')
```

```python
incendios_93_19_projected.plot()
```

```python
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(15)
incendios_93_19_projected.plot(ax=ax, color='white', edgecolor='black')
geo_03102_sample_df.sample(30).plot(ax=ax, marker='o', color='red', markersize=1, aspect=1)
plt.show()
```

```python
points_within_03102 = gpd.sjoin(geo_03102_sample_df, incendios_93_19_projected, predicate='within')
points_within_03102
```

```python
len(points_within_03102.index)
```

```python
geo_03102_sample_fire_df = geo_03102_sample_df.copy()
```

```python
geo_03102_sample_fire_df['is_fire'] = geo_03102_sample_df.index.isin(points_within_03102.index)
```

```python
# Exportamos
geo_03102_sample_fire_df.to_file('data/municipios/datos_generados/03102/geo_03102_fire.geojson', driver='GeoJSON')
```
